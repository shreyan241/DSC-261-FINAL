```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Assuming 'sampler' is already set to "RBF"
sampler = "RBF"

# Load the data
df_original = pd.read_csv(r'C:\Users\shrey\Desktop\DSC 261\DSC-261-FINAL\Data\compas_RBF_train.csv')
df_other = pd.read_csv(rf'C:\Users\shrey\Desktop\DSC 261\DSC-261-FINAL\Data\compas_adversarial_train_{sampler}.csv')
df_lime = pd.read_csv(rf'C:\Users\shrey\Desktop\DSC 261\DSC-261-FINAL\Data\compas_{sampler}.csv')

# One-hot encode all categorical variables
df_original = pd.get_dummies(df_original)
df_original = df_original.drop(columns=['response'])
df_other = pd.get_dummies(df_other)
df_lime = pd.get_dummies(df_lime)

# Assuming all dataframes have the same columns after one-hot encoding
# If not, align the columns using `pd.align` or similar methods

# Combine all the data
X_combined = pd.concat([df_original, df_other, df_lime])

# Standardize the features
X_std = StandardScaler().fit_transform(X_combined)

# Perform PCA
pca = PCA(n_components=2)
principalComponents_combined = pca.fit_transform(X_std)

# Slice the transformed data back into original, other, and lime datasets
num_original = len(df_original)
num_other = len(df_other)
num_lime = len(df_lime)

principalDf_original = pd.DataFrame(data=principalComponents_combined[:num_original], columns=['PCA1', 'PCA2'])
principalDf_other = pd.DataFrame(data=principalComponents_combined[num_original:num_original+num_other], columns=['PCA1', 'PCA2'])
principalDf_lime = pd.DataFrame(data=principalComponents_combined[-num_lime:], columns=['PCA1', 'PCA2'])

# Plot PCA1 vs PCA2 for all datasets
plt.figure(figsize=(8, 6))
plt.scatter(principalDf_original['PCA1'], principalDf_original['PCA2'], color='blue', s=1, alpha=0.3, label='Original Data')
plt.scatter(principalDf_other['PCA1'], principalDf_other['PCA2'], color='red', s=1, alpha=0.3, label=f'{sampler}')
plt.scatter(principalDf_lime['PCA1'], principalDf_lime['PCA2'], color='green', s=1, alpha=0.3, label='LIME')
plt.xlabel('Principal Component 1 (PCA1)')
plt.ylabel('Principal Component 2 (PCA2)')
plt.title('Comparison of PCA between Original, RBF, and LIME')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler

sampler = "forest"
# Load the original data and the other data
df_original = pd.read_csv(rf'C:\Users\shrey\Desktop\DSC 261\DSC-261-FINAL\Data\compas_{sampler}_train.csv')
df_other = pd.read_csv(rf'C:\Users\shrey\Desktop\DSC 261\DSC-261-FINAL\Data\compas_adversarial_train_{sampler}.csv')
df_original = pd.get_dummies(df_original)
df_other = pd.get_dummies(df_other)

# Select the features (assuming both datasets have the same feature columns)
features_original = df_original.columns[:-1]  # Get rid of response

# Standardize the features for both datasets
x_original = StandardScaler().fit_transform(df_original[features_original])
x_other = StandardScaler().fit_transform(df_other)

# Combine the datasets for t-SNE
x_combined = np.vstack([x_original, x_other])
y_combined = np.array(['Original'] * x_original.shape[0] + [sampler] * x_other.shape[0])

# Set perplexity values
perplexities = [5, 25, 50, 100]

# Create a 2x2 plot with t-SNE representations for different perplexity values
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

for i, perplexity in enumerate(perplexities):
    # Perform t-SNE on the combined dataset
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    tsneComponents_combined = tsne.fit_transform(x_combined)

    # Plot t-SNE1 vs t-SNE2 for both datasets
    ax = axs[i // 2, i % 2]
    ax.scatter(tsneComponents_combined[:x_original.shape[0], 0], tsneComponents_combined[:x_original.shape[0], 1], color='blue', label='Original Data', alpha=0.5)
    ax.scatter(tsneComponents_combined[x_original.shape[0]:, 0], tsneComponents_combined[x_original.shape[0]:, 1], color='red', label=f'{sampler}', alpha=0.5)
    ax.set_xlabel('t-SNE Component 1')
    ax.set_ylabel('t-SNE Component 2')
    ax.set_title(f't-SNE with Perplexity = {perplexity}')
    ax.legend()
    ax.grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
